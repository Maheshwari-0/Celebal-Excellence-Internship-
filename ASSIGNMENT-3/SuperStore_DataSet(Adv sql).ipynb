Orders Table

```sql
create Table orders (
 row_id int primary key ,
order_id string not null,
order_date string not null,
ship_date string not null,
ship_mode string not null,
customer_id string not null,
 product_id string not null,
sales double not null,
quantity int not null,
discount double not null,
profit double,

constraint fk_customer
    foreign key(customer_id) references customers(customer_id),

constraint fk_product
    foreign key(product_id) references products(product_id)
);
```

### Step 2: Perform Required Queries


#### 1.Find all orders where sales are greater than the average sales. (Subquery)  

```sql
select * from orders where sales > (select avg(sales) from orders) ;
```

Insight

This query uses a subquery to calculate the average sales value across all orders and then retrieves only those orders whose sales are greater than the overall average

#### 2.Find the highest sales order for each customer. (Subquery)  

```sql
select
customer_id,order_id,sales
from orders o
where sales = (
select MAX(sales)
 from orders where  customer_id = o.customer_id
);
```

Insight
> This query finds the order with the highest sales for each customer, helping identify their most valuable purchase and highlighting high spending customer.

#### 3.Calculate total sales for each customer. (CTE)  

```sql
with CTE_Total_sales as(
select customer_id,
    sum(sales) as Total_sales
    from orders
      group BY customer_id
)
select  *  from CTE_Total_sales;
```

**Insight:**
This query uses a Common Table Expression (CTE) to calculate the total sales generated by each customer. By aggregating sales using SUM(sales) and grouping by customer_id, it provides a clear view of each customer's overall contribution to revenue.

#### 4.Find customers whose total sales are above average. (CTE + Subquery)  

```sql
with customers_Total_sales AS(
select
customer_id,
sum(sales) as Total_sales
from orders 
  group BY customer_id
)
select
* from customers_Total_sales
where Total_sales >(select AVG(Total_sales) from customers_Total_sales);
```

Insight

This query combines a CTE and a subquery to identify customers whose total sales exceed the average sales across all customers. The CTE first calculates the total sales for each customer, while the subquery computes the average of those totals. Customers with sales above this average are considered high-performing customer

#### 5.Rank all customers based on total sales. (Window Function)  

```sql
select *,
RANK() over(Order BY Total_sales DESC)as customer_Rank
 FROM 
 (select
 customer_id,
 sum(sales) as Total_sales
from orders group by customer_id)t;
```

Insight 

This query ranks customers based on their total sales using the RANK() window function. Customers with higher total sales receive better ranks,If multiple customers have the same total sales, they are assigned the same rank.

#### 6.Assign row numbers to each order within a customer. (Window Function + PARTITION BY)  

```sql
select 
order_id,customer_id,
ROW_NUMBER() over(PARTITION BY customer_id  order by order_id) as customer_row_number 
 from orders; 
```

Insight 

This query uses the ROW_NUMBER() window function along with Partition by customer_id to assign a unique sequential number to each order within a customer's order

#### 7.Display top 3 customers based on total sales. (Window Function)  

 

```sql
with customer_sales as (
select
customer_id,
sum(sales) as total_sales
from orders group by customer_id
)
select *
from(
select
customer_id,
total_sales,
RANK() over(order by total_sales DESC) as customer_rank
    from customer_sales
) t
where customer_rank<=3;
```

Insight 

This query identifies the top 3 customers based on their total sales using the RANK() window function. First, a CTE calculates the total sales for each customer. Then customers are ranked in descending order of total sales, and only those with ranks 1 to 3 are displayed.

## Step 3: Final Combined Query 


#### Write one final query that shows:
- Customer Name
- Total Sales
- Rank

(Use JOIN + CTE + Window Function together) 

```sql
with customer_sales as (
select
c.customer_name,
SUM(o.sales) as total_sales
from orders o
join customers c on o.customer_id = c.customer_id
group by c.customer_name
)
select
customer_name,
total_sales,
RANK() over (order by total_sales DESC) as customer_rank
from customer_sales;
```

Insight for Assignment

This query combines a Join, CTE, and Window Function to provide a complete customer sales analysis. The JOIN retrieves customer names from the customer table, the CTE calculates total sales for each customer, and the RANK() window function ranks customers based on their total sales.
