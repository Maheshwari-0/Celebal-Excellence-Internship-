### 3.Handle missing values (identify, fill/drop). 


```python
df.isnull().sum()
```

```python
print("Total missing values:",df.isnull().sum().sum())
```

```python
#Drop the columns that are to empty to be useful
#what_customer_said=573,videos=781,variations=562
df=df.drop(columns=["what_customers_said","videos","variations"])
print("Remaining columns:",df.shape[1])
```

```python
#Filling discount missing values with 0
df["discount"]=df["discount"].fillna(0)
print("Missing values in discount:",df["discount"].isnull().sum())
```

```python
#filling seller_name with unknown
df["seller_name"]=df["seller_name"].fillna('unknown')
print("Missing values in seller_name:",df["seller_name"].isnull().sum())
```

```python
#filling seller_information with Not available
df["seller_information"]=df["seller_information"].fillna("Not Available")
print("Missing values in seller_information:",df["seller_information"].isnull().sum())
```

```python
#checking missing values in dataframe
df.isnull().sum()
```

```python
#filtering products with ratings above 5
high_rated=df[df["rating"]>4.5]
print("proucts with rating above 4.5:",len(high_rated))
high_rated[["title","category","rating"]].head()
```

```python
#filtering specific category(watches)
watches=df[df["category"]=="watches"]
print("Total watch products:",len(watches))
watches[["title","category","rating","final_price"]].head()
```

```python
#filtering using two conditions
#high rated and big discount
best_value=df[(df["rating"]>4.0) & (df["discount"]>50)]
print("High rated+ big discount products:",len(best_value))
best_value[["title","category","rating","discount","final_price"]].head()
```

### 5.Remove duplicates. 


```python
print("Total no of duplicates",df.duplicated().sum())
```

```python
print("Duplicate product_ids",df["product_id"].duplicated().sum())
```

```python
before=len(df)
df=df.drop_duplicates()
after=len(df)
print("Duplicates Removed:",(before-after))
```

### 6.Create a derived column (total_amount = price * quantity). 


```python
#creating quantity column
```

```python
df['quantity']=1
print("quantity column created")
df[["title","final_price","quantity"]].head(10)
```

```python
df['final_price']=df['final_price'].astype(str)
df['final_price']=df['final_price'].str.replace('₹','',regex=False)
df['final_price']=df['final_price'].str.replace(',','',regex=False)
df['final_price']=df['final_price'].str.replace('"','',regex=False)
df['final_price']=df['final_price'].astype(float)
```

```python
df["total_amount"]=df["final_price"]*df["quantity"]
```
