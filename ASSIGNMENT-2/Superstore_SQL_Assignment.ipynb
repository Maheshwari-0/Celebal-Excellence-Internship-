## 4.Use GROUP BY for aggregations (sales, quantity, averages).

#### Query 1:Total Sales and Profit by Region

```sql
SELECT Region,ROUND(SUM(Sales),2) AS Total_Sales,ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore_dataset
GROUP BY Region ORDER BY Total_Sales DESC;
```

Insight:

- Compares sales and profitability across regions.
- Helps identify the best performing market.

#### Query 2:Total Quantity Sold by Category 


```sql
SELECT Category,SUM(Quantity) AS Total_Quantity
FROM superstore_dataset
GROUP BY Category ORDER BY Total_Quantity DESC;
```

Insight:

- Shows which product categories are most in demand.
- Useful for inventory planning.

#### Query 3:Average Sales and Average Profit by Customer Segment 

```sql
SELECT Segment,round(AVG(Sales),2) AS Avg_Sales,round(AVG(Profit),2) AS Avg_Profit
FROM superstore_dataset
GROUP BY Segment ORDER BY Avg_Sales DESC;
```

Insight:

- Reveals which customer segment generates larger orders and profits.
- Supports customer targeting strategies.

#### Query 4:Top 10 States by Total Profit 

```sql
SELECT State,ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore_dataset
GROUP BY State ORDER BY Total_Profit DESC LIMIT 10;
```

Insight:

- Identifies the most profitable states.
- Helps prioritize sales efforts geographically.

#### Query 5:Sales, Profit, and Quantity by Category

```sql
SELECT Category,ROUND(SUM(Sales),2) AS Total_Sales,SUM(Quantity) AS Total_Quantity,
ROUND(SUM(Profit),2) AS Total_Profit FROM superstore_dataset
GROUP BY Category ORDER BY Total_Sales DESC;
```

Insight:

- Provides a complete business view of each category.
- This  identifies which product categories contribute the most to revenue and profit.

Insight:

- Indicates potential issues with pricing, discount policies, or operational costs.

## 6.Solve use cases (monthly trends, top customers, duplicates)

#### Query 1:Monthly Sales Trend 

```sql
SELECT
    YEAR(
        COALESCE(
            try_to_timestamp(`Order Date`, 'MM-dd-yyyy'),
            try_to_timestamp(`Order Date`, 'M/d/yyyy')
        )
    ) AS Year,
    MONTH(
        COALESCE(
            try_to_timestamp(`Order Date`, 'MM-dd-yyyy'),
            try_to_timestamp(`Order Date`, 'M/d/yyyy')
        )
    ) AS Month,
    ROUND(SUM(Sales),2) AS Total_Sales
FROM superstore_dataset
GROUP BY
    YEAR(
        COALESCE(
            try_to_timestamp(`Order Date`, 'MM-dd-yyyy'),
            try_to_timestamp(`Order Date`, 'M/d/yyyy')
        )
    ),
    MONTH(
        COALESCE(
            try_to_timestamp(`Order Date`, 'MM-dd-yyyy'),
            try_to_timestamp(`Order Date`, 'M/d/yyyy')
        )
    )
ORDER BY Year, Month;
```

Insight:
- Tracks sales performance over time and identifies seasonal trends.
- Helps determine peak and low-performing months for business planning.
- Useful for forecasting future demand and revenue.

#### Query 2:Top 10 Customers by Revenue 

```sql
SELECT `Customer Name`,COUNT(DISTINCT `Order ID`) AS Total_Orders,ROUND(SUM(Sales),2) AS Total_Sales,
ROUND(SUM(Profit),2) AS Total_Profit FROM superstore_dataset
GROUP BY `Customer Name` ORDER BY Total_Sales DESC LIMIT 10;
```

Insight:
- Identifies the most valuable customers based on spending and profitability.

#### Query 3:Duplicate Orders

```sql
SELECT `Order ID`,COUNT(*) AS Records_Per_Order
FROM superstore_dataset
GROUP BY `Order ID` HAVING COUNT(*) > 1 ORDER BY Records_Per_Order DESC;
```

##  7.Validate results (row counts, data quality).

#### Query 1:Total Row Count

```sql
SELECT COUNT(*) AS Total_Rows FROM superstore_dataset;
```

Insight
- Confirms that the dataset was loaded successfully.
- Return the Total Number of Rows

#### Query 2:Checking for Missing Values

```sql
SELECT
    SUM(CASE WHEN `Order ID` IS NULL THEN 1 ELSE 0 END) AS Missing_OrderID,
    SUM(CASE WHEN `Customer ID` IS NULL THEN 1 ELSE 0 END) AS Missing_CustomerID,
    SUM(CASE WHEN `Product ID` IS NULL THEN 1 ELSE 0 END) AS Missing_ProductID,
    SUM(CASE WHEN Sales IS NULL THEN 1 ELSE 0 END) AS Missing_Sales,
    SUM(CASE WHEN Profit IS NULL THEN 1 ELSE 0 END) AS Missing_Profit
FROM superstore_dataset;
```

Insight
- Verifies data completeness.
- Missing values in key columns can affect analysis accuracy.

#### Query 3:Checking for Duplicate Rows

```sql
SELECT *, COUNT(*) AS Duplicate_Count
FROM superstore_dataset
GROUP BY ALL HAVING COUNT(*) > 1;
```

Insight
- Detects fully duplicated records.
- Ensures sales and profit are not counted multiple times.

#### Query 4:Checking for Invalid Sales or Quantity


```sql
SELECT *
FROM superstore_dataset WHERE Sales <= 0 OR Quantity <= 0;
```

Insight
- Identifies potentially incorrect transactions.
- Negative or zero sales/quantity values may indicate data entry issues.

#### Query 5:Checking for Orders Shipped Before They Were Placed

```sql
SELECT * FROM superstore_dataset
WHERE `Ship Date` < `Order Date`;
```

Insight
- Verifies the logical consistency of order and shipping dates.
